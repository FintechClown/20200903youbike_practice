```python
import json
import requests
import pandas as pd
import numpy as np
```

```python
r = requests.get("https://tcgbusfs.blob.core.windows.net/blobyoubike/YouBikeTP.json")
ret = r.json()
retVal = ret['retVal']
type(retVal)
```

```python
col_name = 'sno(站點代號)、sna(場站中文名稱)、tot(場站總停車格)、sbi(場站目前車輛數量)、sarea(場站區域)、mday(資料更新時間)、lat(緯度)、lng(經度)、ar(地點)、sareaen(場站區域英文)、snaen(場站名稱英文)、aren(地址英文)、bemp(空位數量)、act(全站禁用狀態)'
col = col_name.split('、')
new_col=[]
for x in range(len(col)):
    new_col.append('') 
#建立一個等長的new_col來儲存純英文名字

for i in range(len(col)):
       for j in range(len(col[i])):
            if col[i][j] in 'abcdefghijklmnopqrstuvwxyz':
                new_col[i] += col[i][j]
#每個 list element裡的每個字元，判斷是否為英文，是則存為 new_col每個 element中的字元
print(new_col)
```

```python
all_list=[]
for i in range(len(new_col)):
    all_list.append([])
```

```python
all_list
```

```python
list_a=[]
```

```python
for i in retVal:
    for j in new_col:     
        list_a.append(retVal[i][j])
```

```python
for i in range(len(all_list)):
    for j in range(i,len(list_a),len(all_list)):
        all_list[i].append(list_a[j])
```
